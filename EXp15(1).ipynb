Part A: Data Normalization

```python
import pandas as pd
import numpy as np

data = {

'Product': ['Laptop', 'Mobile', 'Tablet', 'Headphones', 'Smartwatch', 'Camera'],
 'Price': [55000, 20000, 30000, 2000, 8000, 45000],
'Units_Sold': [25,60,35,80,50,15],
'Discount': [10,15,12,20,18,8]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Min-Max Normalization
# 1. Min-Max Normalization for Price

df['Price_MinMax'] = ((df['Price'] - df['Price'].min()) / (df['Price'].max() - df['Price'].min()))
print(df)
```

```python
# Normalize  multiple colums

cols = ['Price','Units_Sold','Discount']

df_norms = ((df[cols] - df[cols].min()) / (df[cols].max() - df[cols].min()))
print(df_norms)
```

```python
# 2. Z-score Normalization

cols1 = ['Price','Units_Sold','Discount']

df_zscore = ((df[cols1] - df[cols].mean()) / df[cols].std())
print(df_zscore)
```

```python
# 3. Decimal Scaling

df['price_Decimal'] = df['Price'] /  100000
print(df)
```
